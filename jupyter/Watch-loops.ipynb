```python
import matplotlib.animation
from matplotlib import pyplot
import numpy
from IPython.display import HTML
```

```python
arr = numpy.load("output_slices.npz")
x = arr["slices"]
```

```python
fig, ax = pyplot.subplots()
im = ax.imshow(x[0,:,:])

def animate(i):
    im.set_data(x[i,:,:])
ani = matplotlib.animation.FuncAnimation(fig, animate, frames=x.shape[0])

HTML(ani.to_jshtml())
```
