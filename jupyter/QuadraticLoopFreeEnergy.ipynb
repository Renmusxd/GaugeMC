```python
import numpy
from matplotlib import pyplot
```

```python
# arr = numpy.load("output_distribution_k06.npz")
arr = numpy.load("output_distribution_k06_big.npz")
# arr = numpy.load("output_distribution_k10.npz")
sample_probs_06 = arr["sample_probs"]

L = 8
nr = L
```

```python
pyplot.plot(sample_probs_06, 'x-')
pyplot.grid()
pyplot.yscale('log')
pyplot.xlabel("Replica index")
pyplot.ylabel("Replica population fraction")
pyplot.show()

for r in range(nr):
    square_probs_06 = numpy.array([sample_probs_06[(L*L)*r + i**2] for i in range(L)])
    pyplot.plot(square_probs_06, '.-')
pyplot.grid()
pyplot.yscale('log')
pyplot.show()
```

```python
if "free_energies" in arr:
    free_energy = arr["free_energies"]
    all_sqr_free_energy = numpy.array([free_energy[(L*L)*r + i**2] for r in range(nr) for i in range(L)])
else:
    all_sqr_probs = numpy.array([sample_probs_06[(L*L)*r + i**2] for r in range(nr) for i in range(L)])
    all_sqr_free_energy = -numpy.log(all_sqr_probs) + numpy.log(all_sqr_probs[0])

pyplot.plot(all_sqr_free_energy, 'x-')
pyplot.scatter([L*r for r in range(nr)], [all_sqr_free_energy[L*r] for r in range(nr)], c='r')

pyplot.ylabel("Free energy cost")
pyplot.xlabel("Wilson loop 'index'")
pyplot.grid()
pyplot.show()
```

```python
pyplot.plot([all_sqr_free_energy[L*r] for r in range(nr)], 'x-')
pyplot.xlabel("Number of planes")
pyplot.ylabel("Free energy cost")
pyplot.grid()
pyplot.show()

pyplot.plot([r**2 for r in range(nr)], [all_sqr_free_energy[L*r] for r in range(nr)], 'x-')
pyplot.xlabel("(Number of planes)^2")
pyplot.ylabel("Free energy cost")
pyplot.grid()
pyplot.show()
```
