```python
import numpy
from matplotlib import pyplot
```

```python
arr = numpy.load("output_distribution_k05.npz")
sample_probs_05 = arr["sample_probs"]

arr = numpy.load("output_distribution_k15.npz")
sample_probs_15 = arr["sample_probs"]
```

```python
pyplot.plot(sample_probs_05, 'x-')
pyplot.plot(sample_probs_15, 'x-')
pyplot.grid()
pyplot.yscale('log')
pyplot.show()
```

```python
square_probs_05 = numpy.array([sample_probs_05[i**2] for i in range(16+1)])
square_probs_15 = numpy.array([sample_probs_15[i**2] for i in range(16+1)])
pyplot.plot(numpy.arange(16+1), square_probs_05, 'x-')
pyplot.plot(numpy.arange(16+1), square_probs_15, 'x-')
pyplot.grid()
pyplot.show()
```

```python
pyplot.plot(numpy.arange(16+1), -numpy.log(square_probs_05) + numpy.log(square_probs_05[0]), 'x-')
pyplot.plot(numpy.arange(16+1), -numpy.log(square_probs_15) + numpy.log(square_probs_15[0]), 'x-')
pyplot.grid()
pyplot.show()
```

```python
pyplot.plot(numpy.arange(16+1)**2, -numpy.log(square_probs_05) + numpy.log(square_probs_05[0]), 'x-')
pyplot.plot(numpy.arange(16+1)**2, -numpy.log(square_probs_15) + numpy.log(square_probs_15[0]), 'x-')
pyplot.grid()
pyplot.show()
```
